# 1.1 Grafiklere Argumen verme

```python
x = np.linspace(0, 10, 100) ## 0 ile 10 arasında 100 adet eşit aralıklı sayı üretir

y = np.sin(x)



plt.figure(figsize = (8,6))
plt.plot(x, y,  "r*")  # p yi marker olarak algiladi

plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(x, y, 'g--', linewidth=2, markersize=12, );
```

```python
x = np.linspace(0, 10, 100) 
y = np.sin(x)


plt.figure(figsize = (10,5))

plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.plot(x, y,  "*g") 
plt.title(label = "AXES NAME")
plt.show()
```

# Coklu Data ve Legend Verme - Plotting multiple sets of Data


```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000,76000, 5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83454]
```

```python
plt.figure(figsize = (10,5))
plt.plot(age, salary,"b", label = "SALARY")
plt.plot(age, salary_2,"r", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY by AGE")     

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(age, salary,"--", label = "SALARY")
plt.plot(age, salary_2,"r", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY by AGE")     

plt.legend(loc = "lower right") ##legend yeri degistirmek icin
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(age, salary,"--", label = "SALARY")
plt.plot(age, salary_2,"r", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY by AGE")     

plt.legend(loc = 5) ##legend yeri degistirmek icin
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.plot(age, salary,"--", label = "SALARY")
plt.plot(age, salary_2,"r", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY by AGE")     

plt.legend(bbox_to_anchor= (0,0,0,0)) ##ince ayarlar yapmak icin
plt.show()
```

# 1.3 Figure and Axes Renklendirmesi

```python
plt.figure(facecolor="pink", figsize = (10, 5))  #figurumu cercevesini renklendirdim

ax = plt.axes() #grafigimi ax e atamis oldum

ax.set_facecolor("xkcd:mint green") # rengimi mint green e set ettim


plt.plot(age, salary, "b-", linewidth = 4, label = "SALARY 1") 
plt.plot(age, salary_2, "r", linestyle  = ":", label = "SALARY 2") #cizginin stili

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title(label = "SALARY BY AGE") #baslik adi


plt.legend(bbox_to_anchor = (0.5,0.7, 0.5, 0.5))
plt.show()
```

```python
plt.figure(facecolor="yellow", figsize = (10, 5))  #figurumu cercevesini renklendirdim

ax = plt.axes() #grafigimi ax e atamis oldum

ax.set_facecolor("xkcd:mint green") # rengimi mint green e set ettim


plt.plot(age, salary, "b-", lw = 6, label = "SALARY 1") #line kalinligi
plt.plot(age, salary_2, "rv", ls  = ":", label = "SALARY 2") #cizginin stili

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title(label = "SALARY BY AGE") #baslik adi


plt.legend(loc = "upper center")
plt.show()
```

# Functional Method

```python
plt.subplot(2,1,1) # 2 satir 1 sutundan olusan bir grafik olusturur. 1. satir
plt.plot(age, salary, "b")



plt.subplot(2,1,2) # 2. satir
plt.plot(age, salary_2, "r");

```

```python
plt.subplot(1,2,1) # 2 satir 1 sutundan olusan bir grafik olusturur. 1. satir
plt.plot(age, salary, "b")



plt.subplot(1,2,2) # 2. satir
plt.plot(age, salary_2, "r");

```

```python
plt.figure(figsize = (12,5)) #Grafik boyunu ayarladik

plt.subplot(1,2,1) 
plt.plot(age, salary, "b")



plt.subplot(1,2,2) 
plt.plot(age, salary_2, "r");
```

```python
plt.figure(figsize = (20,5))

plt.subplot(2, 3, 1)
plt.plot(age, salary, "b")

plt.subplot(2, 3, 2)
plt.plot(age, salary, "ro")

plt.subplot(2, 3, 3)
plt.plot(age, salary, "--")

plt.subplot(2, 3, 4)
plt.plot(age, salary_2, "g")

plt.subplot(2, 3, 5)
plt.plot(age, salary_2, "p")

plt.subplot(2, 3, 6)
plt.plot(age, salary_2, "r");
```

```python

plt.subplot(2, 3, 1)
plt.plot(age, salary, "b")

plt.subplot(2, 3, 2)
plt.plot(age, salary, "ro")

plt.subplot(2, 3, 3)
plt.plot(age, salary, "--")

plt.subplot(2, 3, 4)
plt.plot(age, salary_2, "g")

plt.subplot(2, 3, 5)
plt.plot(age, salary_2, "p")

plt.subplot(2, 3, 6)
plt.plot(age, salary_2, "r");

plt.tight_layout() #Fig size i otomatik ayarladi ve grafigin birbirine gecmesini otomatik engellemis oldu
plt.show()
```

# Object Oriented Method (OOM)



```python
fig, ax = plt.subplots(2,1)

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout() #grifiğin birbirine geçmesini engelliyor
plt.show()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")
plt.tight_layout() #grifiğin birbirine geçmesini engelliyor
plt.show()
```

```python
fig, ax = plt.subplots(2, 3) #grafigim 2 satir 3 sutundan olussun

ax[0][0].plot(age, salary, "b")
ax[0][1].plot(age, salary, "c")
ax[0][2].plot(age, salary, "b")


ax[1][0].plot(age, salary_2, "r")
ax[1][1].plot(age, salary_2, "m")
ax[1][2].plot(age, salary_2, "r")
plt.tight_layout() 
plt.show() 
```
