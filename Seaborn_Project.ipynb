## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill["manner_of_death"].unique()
```

#### INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
kill.dropna(inplace=True)
```

```python
plt.figure(figsize=(4.5,3))
sns.boxplot(x = "gender", y ="age", data = kill, hue = "manner_of_death");
```

```python
# with seaborn
```

## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
plt.figure(figsize=(4.5,3))
sns.swarmplot(data= kill, x= "gender", y= "age", hue="manner_of_death")
plt.show();
```

## 2: Plot 12 most dangerous cities

```python
kill["city"].value_counts().head(12)
```

```python
a=kill["city"].value_counts().head(12)
y = a.values
y
```

```python
x = a.index
x
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x=x,y=y)
plt.title(label = "Most Dangerous Cities" , color = "red")
plt.xticks(rotation=45, color = "blue", fontsize= 8)
plt.xlabel("")
plt.tight_layout()
```

```python
# with seaborn
```

```python
kill["city"].value_counts().head(12)
```

```python
a=kill["city"].value_counts().head(12)
a.index
```

```python
plt.figure(figsize=(3,2))
top_12_cities = kill['city'].value_counts().head(12)
plt.figure(figsize=(10, 6))
sns.countplot(x='city', data=kill, order=top_12_cities.index)
plt.title('Most Dangereous Cities', color = "blue")
plt.xticks(rotation=45, color= "black")
plt.show();
```

```python
# with seaborn countplot
```
